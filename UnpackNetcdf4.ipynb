```python
import netCDF4
import pandas as pd

precip_nc_file = 'iwv.nc'
nc = netCDF4.Dataset(precip_nc_file, mode='r')

nc.variables.keys()
```

```python
lat = nc.variables['latitude'][:]
lon = nc.variables['longitude'][:]
time_var = nc.variables['time']
dtime = netCDF4.num2date(time_var[:],time_var.units)
air = nc.variables['p55.162'][:]
```

```python
import numpy as np
cek = pd.DataFrame()
for i in air:
    cek = cek.append(pd.Series(i[0]), ignore_index=True)
```

```python
cek.index = dtime
cek.head()
```
